```python
def read_file_in_df(filepath):
    x = []; y = [];
    with open(filepath, 'r') as fp:
        for line in fp:
            p, r = line.split(',')
            p = float(p.strip()); r = float(r.strip());
            if r < 1:
                continue
            x.append(p); y.append(r);
    return x, y
```

```python
ep_greedy_x, ep_greedy_y = read_file_in_df('./stationary/ep_greedy_smpl.csv')
```

```python
gradient_bandit_x, gradient_bandit_y = read_file_in_df('./stationary/gradient_bandit.csv')
```

```python
ucb_x, ucb_y = read_file_in_df('./stationary/ucb.csv')
```

```python
optimistic_greedy_x, optimistic_greedy_y = read_file_in_df('./stationary/opti_greedy.csv')
```

```python
plt.plot(ep_greedy_x, ep_greedy_y, 'r-')
plt.plot(gradient_bandit_x, gradient_bandit_y, 'g-')
plt.plot(ucb_x, ucb_y, 'b-')
plt.plot(optimistic_greedy_x, optimistic_greedy_y, 'k')
plt.xscale('log')
```
